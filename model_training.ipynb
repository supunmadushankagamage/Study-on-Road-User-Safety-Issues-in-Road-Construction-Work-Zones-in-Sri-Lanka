```python
unique_factors = set()

processed_dfs = []

for i in range(len(dfs)):
  df = dfs[i]

  df = df.rename(columns={"Exposure" : "Exposure_High", "Unnamed: 2": "Exposure_Low", "Unnamed: 3": "Exposure_Not_Apply"})
  df = df.rename(columns={"Severity" : "Severity_Major", "Unnamed: 5": "Severity_Minor", "Unnamed: 6": "Severity_Not_Apply"})
  df = df.rename(columns={"Likelihood": "Likelihood_High", "Unnamed: 8": "Likelihood_Low", "Unnamed: 9" : "Likelihood_Not_Apply"})

  df = df[["Factors", "Exposure_High", "Exposure_Low", "Exposure_Not_Apply", "Severity_Major", "Severity_Minor", "Severity_Not_Apply", "Likelihood_High", "Likelihood_Low", "Likelihood_Not_Apply"]]

  df = df.dropna()

  unique_values = df["Factors"].unique()
  print(f"number of unique factors: {len(unique_values)} for dataframe {i}")
  unique_factors = unique_factors.union(set(unique_values))
  print(f"number of unique factors: {len(unique_factors)} for all dataframes so far")

  column_names = df.columns
  # print(f"column names: {column_names}")

  df = df.replace({"TRUE": True, "FALSE": False})

  # df["Factors"] = df["Factors"].apply(lambda x: x.strip())

  df["Exposure"] = df[["Exposure_High", "Exposure_Low", "Exposure_Not_Apply"]].any(axis=1)
  df["Severity"] = df[["Severity_Major", "Severity_Minor", "Severity_Not_Apply"]].any(axis=1)
  df["Likelihood"] = df[["Likelihood_High", "Likelihood_Low", "Likelihood_Not_Apply"]].any(axis=1)

  processed_dfs.append(df)
```

```python
unique_factors = list(unique_factors)
unique_factors.sort()
print(unique_factors)
print(len(unique_factors))
```

```python
final_dfs = []
for processed_df in processed_dfs:
  # Find missing values
  existing_values = set(processed_df['Factors'])
  missing_values = [val for val in unique_factors if val not in existing_values]
  print(existing_values)
  print(missing_values)

  # Append missing rows
  for val in missing_values:
      processed_df = pd.concat([processed_df, pd.DataFrame({'Factors': [val], 'Exposure_High': [False], 'Exposure_Low': [False], 'Exposure_Not_Apply': [False],
       'Severity_Major': [False], 'Severity_Minor': [False], 'Severity_Not_Apply': [False],
       'Likelihood_High': [False], 'Likelihood_Low': [False], 'Likelihood_Not_Apply': [False], 'Exposure': [False], 'Severity': [False], 'Likelihood': [False]})], ignore_index=True)

  processed_df["Exists"] = processed_df[["Exposure", "Likelihood", "Severity"]].any(axis=1)
  final_dfs.append(processed_df)

final_unique_factors = set()
for df in final_dfs:
  unique_values = df["Factors"].unique()
  print(f"number of unique factors: {len(unique_values)} for dataframe {i}")
  final_unique_factors = final_unique_factors.union(set(unique_values))
  print(f"number of unique factors: {len(final_unique_factors)} for all dataframes so far")
```

## DecisionTreeClassifier

```python
# create X
X_1 = []
X_2 = []
for i in range(len(final_dfs)):
  df = final_dfs[i]
  df = df.sort_values(by='Factors').reset_index(drop=True)
  sample_1 = []
  sample_2 = []
  for index, row in df.iterrows():
    if row['Exposure']:
      if row['Exposure_High'] or row['Exposure_Low']:
        sample_2.append(1)
      else:
        sample_2.append(0)
    else:
      sample_2.append(0)

    if row['Severity']:
      if row['Severity_Major'] or row['Severity_Minor']:
        sample_2.append(1)
      else:
        sample_2.append(0)
    else:
      sample_2.append(0)

    if row['Likelihood']:
      if row['Likelihood_High'] or row['Likelihood_Low']:
        sample_2.append(1)
      else:
        sample_2.append(0)
    else:
      sample_2.append(0)

    for row_name in ['Exposure_High', 'Exposure_Low', 'Exposure_Not_Apply', 'Severity_Major', 'Severity_Minor', 'Severity_Not_Apply', 'Likelihood_High', 'Likelihood_Low', 'Likelihood_Not_Apply']:
      if row[row_name]:
        sample_1.append(1)
      else:
        sample_1.append(0)

  # print(sample_1)
  # print(sample_2)
  X_1.append(sample_1)
  X_2.append(sample_2)
```

```python
# Your dataset
y = ["High", "High", "Medium", "Low", "High", "Medium", "High", "Medium", "Medium", "Low", "Medium", "High", "Low", "Low", "Low", "Low", "Low", "Medium", "Low", "Low", "Low", "Low", "Low", "Medium", "Medium", "Medium", "Low", "Low", "High", "Low", "High", "High", "High", "High", "Medium", "Low", "Low", "Low", "Medium", "Low"]
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score
import numpy as np

for scoring in ['accuracy', 'precision_macro', 'recall_macro', 'f1_macro']:
  print(f"Scoring: {scoring} ===========================================================================")
  # Initialize model
  clf_1 = DecisionTreeClassifier(random_state=42)

  # Perform 5-fold cross-validation
  scores_1 = cross_val_score(clf_1, X_1, y, cv=5, scoring=scoring)

  # Output the accuracy scores for each fold and the average
  print("Cross-validation scores for each fold:", scores_1)
  print("Average accuracy:", np.mean(scores_1))
  print()
```

```python
for scoring in ['accuracy', 'precision_macro', 'recall_macro', 'f1_macro']:
  print(f"Scoring: {scoring} ===========================================================================")
  # Initialize model
  clf_2 = DecisionTreeClassifier(random_state=42)

  # Perform 5-fold cross-validation
  scores_2 = cross_val_score(clf_2, X_2, y, cv=5, scoring=scoring)

  # Output the accuracy scores for each fold and the average
  print("Cross-validation scores for each fold:", scores_2)
  print("Average accuracy:", np.mean(scores_2))
  print()
```

```python
from xgboost import XGBClassifier
from sklearn.preprocessing import LabelEncoder
import warnings
warnings.filterwarnings("ignore")

# Encode string labels to integers
le = LabelEncoder()
y = le.fit_transform(y)  # This will map: 'High'->0, 'Low'->1, 'Medium'->2

# Loop through different scoring metrics
for scoring in ['accuracy', 'precision_macro', 'recall_macro', 'f1_macro']:
    print(f"Scoring: {scoring} ===========================================================================")

    # Initialize XGBoost classifier
    clf_1 = XGBClassifier(use_label_encoder=False, eval_metric='mlogloss', random_state=42)

    # Perform 5-fold cross-validation
    scores_1 = cross_val_score(clf_1, X_1, y, cv=5, scoring=scoring)

    # Output the scores for each fold and the average
    print("Cross-validation scores for each fold:", scores_1)
    print(f"Average {scoring}:", np.mean(scores_1))
    print()
```

```python
import zipfile
import os

def zip_folder(folder_path, output_path):
    with zipfile.ZipFile(output_path, 'w', zipfile.ZIP_DEFLATED) as zipf:
        for root, _, files in os.walk(folder_path):
            for file in files:
                file_path = os.path.join(root, file)
                arcname = os.path.relpath(file_path, start=folder_path)
                zipf.write(file_path, arcname)

# Example usage:
zip_folder('models', 'models.zip')
```
